# Scrapping Mutlimedia Files Using Beautiful Soup



```python
#import modules
from bs4 import BeautifulSoup as bs
import requests
```

```python
#get the page content from url
url = "https://economictimes.indiatimes.com/news/india/some-things-to-know-about-radhika-merchant/radhika-merchant/slideshow/92052992.cms"
page = requests.get(url)

# the image page url
```

```python
page
```

```python
page.content
```

```python
soup = bs(page.content,'html.parser')
print(soup.prettify())
```

```python
#find image source link
#one link - find, many - find_all
# img_tag = soup.find('tag name') 
img_tag = soup.find('link') 
img_tag
```

```python
img_tag['href']
```

```python
img_url = img_tag['href']
```

```python
# download the image from url
image = requests.get(img_url)
```

```python
# store this image in a file
with open('image.png','wb') as file:
    for chunk in image.iter_content(chunk_size = 1024): #1024 = 1kb
        file.write(chunk)
        
#iter_content method allows us to specify the 
#size of data to download by defining the chunk-size in bytes. In this case, it's set at 1024 bytes.
```

# Download Youtube Vedios Using Beautiful Soup


```python


vedio = requests.get("https://www.youtube.com/watch?v=UQOVhj8PQCc")

with open('song.mp4','wb') as file:
    for chunk in vedio.iter_content(chunk_size = 1024): #1024 = 1kb
        file.write(chunk)
    #  now you can find the saved vedio in the same folder where this document is stored
```
