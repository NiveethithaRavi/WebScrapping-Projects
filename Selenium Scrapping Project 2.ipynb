# Scrapping 250 Countries and Their Population Using Selenium.



```python
from selenium import webdriver
import pandas as pd # to structure it
```

```python
# set path for chrome driver
# path of our exe file in c drive
path = 'c:\\Chromedriver.exe'
```

```python
browser = webdriver.Chrome(executable_path = path)
```

```python
url = "http://www.scrapethissite.com/pages/simple/"
```

```python
# open the page url in chrome
browser.get(url)
```

```python
#get country names
from selenium.webdriver.common.by import By
```

```python
country_list = browser.find_elements(By.XPATH,"(//h3[@class='country-name'])")
country_list
```

```python
#parse the data
countries=[]
for country in country_list:
    # get the text data
    temp = country.text
    countries.append(temp)
    
countries
```

```python
# population list of the country
pop_list = browser.find_elements(By.XPATH,"(//span[@class='country-population'])")
pop_list
```

```python
# parse the data

populations = []
for population in pop_list:
    # get the text data
    temp = population.text
    populations.append(temp)
populations
```

```python
# structuring the data

data = pd.DataFrame()
data['Country Name'] = countries
data['Population'] =  populations
```

```python
data
```

```python
data.to_csv("country_name.csv")
```
